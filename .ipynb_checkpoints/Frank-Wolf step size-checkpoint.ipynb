```python
import scipy
import copt as cp
import numpy as np
from scipy import linalg
from tqdm import tqdm
from sklearn.datasets import load_svmlight_file
import plotly.graph_objects as go
from time import perf_counter
```

```python
EPS = np.finfo(np.float32).eps

def calc_gamma_DR(
    x,
    f,
    direction,
    gap,
    lipschitz_t,
    eta=0.9,
    tau=2.0,
    max_step_size=1,
    max_iter=200):

    Qt = lambda gamma, M: - gamma*gap + (gamma**2 * M / 2) * linalg.norm(direction)**2
    f_t = f(x)

    lipschitz_t = eta * lipschitz_t
    if (lipschitz_t * linalg.norm(direction)**2) != 0:
        gamma_t = min(gap / (lipschitz_t * linalg.norm(direction)**2), 1)
    else:
        gamma_t = 1
     
    for _ in range(max_iter):
        if f(x + gamma_t * direction) - f_t <=  Qt(gamma_t, lipschitz_t):
            break
        else:
            lipschitz_t *= tau

    return gamma_t, lipschitz_t

def calc_gamma_adaptive(
    x,
    f_grad,
    direction,
    gap,
    lipschitz_t,
    eta=0.9,
    tau=2.0,
    max_step_size=1,
    max_iter=200):

    lipschitz_t = eta * lipschitz_t
    
    for _ in range(max_iter):
        gamma_t = min(gap / (lipschitz_t * linalg.norm(direction)**2), 1)
        if np.dot(direction, -f_grad(x + gamma_t*direction)) >= 0:
            break
        else:
            lipschitz_t *= tau

    return gamma_t, lipschitz_t

def calc_gamma_search(
    x,
    f,
    direction,
    gamma_t):

    eaxt_line_search = lambda gamma: f(x + gamma*direction)
    gamma = scipy.optimize.minimize(eaxt_line_search, gamma_t, bounds=[(0, 1)], method='L-BFGS-B', options={'ftol': 1e-8, 'gtol': 1e-8}).x[0]

    return gamma

    
    


def FW(
        x0,
        calc_f,
        calc_grad,
        calc_lmo,
        method,
        lipschitz=None,
        n_iter=1000,
        x_sol=None
    ):

    logs = {}

    x = x0.copy()
    lipschitz_t = None
    old_f_t = None

    start = perf_counter()
    
    for t in tqdm(range(1, n_iter+1)):
        grad = calc_grad(x)
        s = calc_lmo(grad, x)
        direction = s-x
        gap = np.dot(direction, -grad)
        
        if lipschitz_t is None:
            eps = 1e-3
            lipschitz_t = linalg.norm(grad - calc_grad(x + eps * direction)) / (eps * linalg.norm(direction) )

        if method == 'backtracing':
            gamma_t, lipschitz_t = calc_gamma_DR(x, calc_f, direction, gap, lipschitz_t)
        elif method == 'adaptive':
            gamma_t, lipschitz_t = calc_gamma_adaptive(x, calc_grad, direction, gap, lipschitz_t)
        elif method == 'line-search':
            gamma_t = calc_gamma_search(x, calc_f, direction, 2 / (t + 2))
        elif method == 'DR':
            assert lipschitz is not None
            gamma_t = min(gap / (lipschitz * linalg.norm(direction))**2, 1)
        else:
            gamma_t = 2 / (t + 2)
        
        logs[perf_counter() - start] = {
            'f_x': calc_f(x),
            'fw': grad @ -direction,
            'dual': grad @ (x - x_sol),
            'primal': calc_f(x) - calc_f(x_sol),
            'gamma': gamma_t
        }
        
        x += gamma_t * direction

    return x, logs
```

## Mushrooms

```python
data = load_svmlight_file("data/mushrooms.txt")
X, y = data[0].toarray(), data[1]
y = 2 * y - 3

_, d = X.shape
```

```python
def f(w, X=X, y=y):
    return np.mean(np.log(1 + np.exp(-y * (X @ w))))

def grad(w, X=X, y=y, return_loss=False):
    n, _ = X.shape
    grad = -(1/n) * (X.T @ (y / (1 + np.exp(y * (X @ w))) ))

    return grad

def lmo(g, x=None, R=1):
    z = np.zeros(len(g))
    
    i = np.argmax(abs(g))
    g_i = g[i]

    if g_i > 0: 
        z[i] = -1*R
    else:
        z[i] = R

    return z
```

```python
calc_f = lambda w: f(w)
calc_grad = lambda w: grad(w)
calc_lmo = lambda grad, x: lmo(grad, x, R=1)
L = 5.25

results = {}

for R in [1, 10, 100]:
    calc_lmo = lambda grad, x: lmo(grad, x, R=R)
    results[R] = {}
    for method in ['simple', 'DR', 'backtracing', 'adaptive', 'line-search']:
        w = np.zeros(d)
        w_sol = scipy.optimize.minimize(calc_f, w, method='BFGS').x
        w, logs = FW(w, calc_f, calc_grad, calc_lmo, method=method, lipschitz=L, n_iter=1000, x_sol=w_sol)
        
        print(f'{method}) f(x)-f(x^*): {calc_f(w)-calc_f(w_sol)}; f(x): {calc_f(w)}')
        results[R][method] = logs
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def plot_nested_dict_plotly_grouped(data, metric):
    """
    Generate line plots (subplots) for a specified metric from a nested dictionary using Plotly.
    Group methods into a single legend.

    Args:
        data (dict): Nested dictionary with structure {param: {method: {x_value: {'metric1': value}}}}
        metric (str): The metric to plot (e.g., 'f_x', 'fw', 'dual', 'primal', 'gamma').
    """
    # Extract top-level keys (parameters)
    params = list(data.keys())
    num_params = len(params)

    # Create subplots
    fig = make_subplots(
        rows=num_params, cols=1,
        subplot_titles=[f"Constraint: {param}" for param in params]
    )

    # Define a set of unique method names to use as grouped legends
    method_colors = {
        'simple': 'blue',
        'DR': 'red',
        'backtracing': 'green',
        'adaptive': 'purple',
        'line-search': 'orange'
    }

    # Iterate over parameters (subplots)
    for idx, param in enumerate(params, start=1):
        methods = data[param]  # Extract methods for the current parameter

        # Add a line for each method
        for method, x_values in methods.items():
            x = list(x_values.keys())
            y = [x_values[x_val][metric] for x_val in x]

            # Add trace with legendgroup to group legends for the same method
            fig.add_trace(
                go.Scatter(
                    x=x, y=y, mode='lines+markers',
                    name=method if idx == 1 else None,  # Add legend only for the first parameter
                    legendgroup=method,  # Group by method name
                    showlegend = True if idx == 1 else False,
                    line=dict(color=method_colors[method]),
                    marker=dict(size=6),
                ),
                row=idx, col=1
            )

        # Set log scale for the x-axis
        fig.update_xaxes(type="log", title_text="Iteration (log scale)", row=idx, col=1)
        fig.update_yaxes(title_text='FW Gap', row=idx, col=1)

    # Update layout
    fig.update_layout(
        title_text=f"Cancer dataset",
        height=800, width=1000,
        showlegend=True,
        template="plotly_white"
    )

    # Show the interactive plot
    fig.show()


plot_nested_dict_plotly_grouped(mushrooms_results, metric="fw")
```

# Datasets

```python
datasets = [
    ("Gisette", cp.datasets.load_gisette, 6e-3),
    ("RCV1", cp.datasets.load_rcv1, 2e4),
    ("Madelon", cp.datasets.load_madelon, 20.0),
    ("Covtype", cp.datasets.load_covtype, 200.0),
]

results = {}

for dataset_title, load_data, alpha in datasets:
    print(dataset_title)
    results[dataset_title] = results.get(dataset_title, {})
    X, y = load_data()
    _, d = X.shape

    l1_ball = copt.constraint.L1Ball(alpha)
    f = copt.loss.LogLoss(X, y)

    calc_f = lambda w: f(w)
    calc_grad = lambda w: f.f_grad(w)[1]
    calc_lmo = lambda grad, x: l1_ball.lmo(-grad, x)[0] + x
    L = f.lipschitz
    n_iter = 400

    for method in ['backtracing', 'line-search', 'simple', 'DR']:
        w = np.zeros(d)
        w_sol = np.zeros(d)
        w, logs = FW(w, calc_f, calc_grad, calc_lmo, method=method, lipschitz=L, n_iter=n_iter, x_sol=w_sol)
        results[dataset_title][method] = logs.copy()
```

```python
plot_nested_dict_plotly_grouped(results, metric="fw")
```
